## ML model Creating,Fitting,scaliing

```python
X=df.drop("diagnosis",axis=1)
X.head()
```

```python
y=df["diagnosis"]
y.head()
```

```python
# spliiting data 75:25 for training(75) ,testing(25) model
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25) # test_size telling about test data (25%) so it it implis train data to 75% so it is not needed to tell train_size seperatly
```

```python
df.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
X_train.head(1)
```

```python
# scaling the  data i.e getting all values to near to 0
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
X_train
```

## ML Model

## 1. Logistic Regression

```python
 from sklearn.linear_model import LogisticRegression
 lr=LogisticRegression() # defing model
 lr.fit(X_train,y_train) # fitting data into model
```

```python
y_pred=lr.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
lr_acc=accuracy_score(y_test,y_pred)
lr_acc
```

```python
results=pd.DataFrame()
results
```

```python
 temp_results=pd.DataFrame({"Algorithm":["Logistic Regression Method"],"Accuracy":[lr_acc]})
 results=pd.concat([results,temp_results])
 results
```

## 2. Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
y_pred=dtc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
dtc_acc=accuracy_score(y_test,y_pred)
dtc_acc
```

```python
 temp_results=pd.DataFrame({"Algorithm":["Decision Tree Classifier Method"],"Accuracy":[dtc_acc]})
 results=pd.concat([results,temp_results])
 results
```

## 3. Random  Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(X_train,y_train)
```

```python
y_pred=rfc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
rfc_acc=accuracy_score(y_test,y_pred)
rfc_acc
```

```python
 temp_results=pd.DataFrame({"Algorithm":["Random Forest Classifier Method"],"Accuracy":[rfc_acc]})
 results=pd.concat([results,temp_results])
 results
```

## 4. Support Vector Classifier

```python
from sklearn import svm
svc=svm.SVC()
svc.fit(X_train,y_train)
```

```python
y_pred=svc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
svc_acc=accuracy_score(y_test,y_pred)
svc_acc
```

```python
temp_results=pd.DataFrame({"Algorithm":["Support Vector Classifier Method"],"Accuracy":[svc_acc]})
results=pd.concat([results,temp_results])
results
```
